```python
import searoute as sr
# whatever destination, here I take the example of the README
destination = [117.42187500000001, 39.36827914916014]

# define your new port
my_nodes = {
    (-77.22, 2.03): {'x': -77.22, 'y': 2.03, "cty": "Colombia", "name": "Balboa", "port": "COBAL", "t": 0.0, "to_cty": ["CO"]},
}
# update new ports into the main list of the Ports
for node in my_nodes.items():
    node_id, attr_node = node
    sr.setup_P().add_node(node_id, **attr_node )

# 
route = sr.searoute((-77.22, 2.03), destination,  include_ports=True)
route
```
